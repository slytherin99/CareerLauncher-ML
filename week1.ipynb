TASK 1.5

```python

def avgprice(N, duration):
    if(duration== 'months' or duration== 'month'):
        n= N*30
    elif(duration== 'years' or duration== 'year'):
        n=N*365
    elif(duration== 'weeks' or duration== 'week'):
        n= N*7
    else:
        n= N
        
    avg= df['Average Price'].tail(n).mean()
    print("Average stock price in the last {n} {duration} = {avg}".format(n= N, duration= duration, avg= avg))
    
    return 
```

```python
def profitloss(N, duration):
    if(duration== 'months' or duration== 'month'):
        n= N*30
    elif(duration== 'years' or duration== 'year'):
        n=N*365
    elif(duration== 'weeks' or duration== 'week'):
        n= N*7
    else:
        n= N
    row=494-n
    ClosePrice_nthday= df.iloc[row]['Close Price'] 
    ClosePrice_lastday= df.iloc[-1]['Close Price'] 
    diff=ClosePrice_lastday-ClosePrice_nthday
    gain_or_loss= (diff/ClosePrice_nthday)*100
    if(diff>=0):
        print("Profit percentage in the last {n} {duration} = {perc}%".format(n= N, duration= duration, perc= gain_or_loss))
        
    else:
        print("Loss percentage in the last {n} {duration} = {perc}%".format(n= N, duration= duration, perc= -gain_or_loss)) 

    return 
```

```python
for i in range(0,6):
    N= int(input())
    duration= str(input())

    avgprice(N, duration)
    profitloss(N, duration)
```

```python
df['Day_Perc_Change'] = df['Close Price'].pct_change(periods = 1)
df.head()
```

TASK 1.7

```python
for i in df['Day_Perc_Change']:
    if i >= -0.5 and i <= 0.5:
        df['Trend'] = 'Slight or No change'
    elif i >= 0.5 and i <= 1:
        df['Trend'] = 'Slight Positive'
    elif i >= -1 and i <= -0.5:
        df['Trend'] = 'Slight Negative'
    elif i >= 1 and i <= 3:
        df['Trend'] = 'Positive'
    elif i >= -3 and i <= -1:
        df['Trend'] = 'Negative'
    elif i >= 3 and i <= 7:
        df['Trend'] = 'Among Top Gainers'   
    elif i >= -7 and i <= -3:
        df['Trend'] = 'Among Top Losers'  
    elif i >7:
        df['Trend'] = 'Bull Run'
    elif i < -7:
        df['Trend'] = 'Bear Drop'
```
